#comparison between CuDF and Pandas through multi bar chart

## this visualization is measured by using User CPU time. which means, the amount of time (miliseconds) the processor spends in running your application code

load data .csv pandas : 20700ms

Using Isna function pandas :  4530ms

Using fillna function pndas : 4630ms



---



load data .csv CUDF : 1960ms

Using Isna function CUDF :29.6 ms

Using fillna function CUDF : 189 ms




```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python

  
X = ['Load data','Fillna function']
Ypandas = [20700,4630 ]
Zcudf = [1960, 189]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.001, Ypandas, 0.8, label = 'Pandas')
plt.bar(X_axis + 0.001, Zcudf, 0.8, label = 'CuDF')
  
plt.xticks(X_axis, X)
plt.xlabel("Library")
plt.ylabel("Miliseconds(ms)")
plt.title("Measure performance based on CPU Time User")
plt.legend()
plt.show()
```

Based on this barchart, we can see that CUDF has faster performance measure for two components, which are load .csv data and using Fillna Function. CuDF and Pandas gain 1960ms,189ms and 20700ms,4630ms , for loading data and using fillna fuction respectively

```python
  
X = ['Isna function ']
Ypandas = [ 4530  ]
Zcudf = [  29.6 ]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.001, Ypandas, 0.8, label = 'Pandas')
plt.bar(X_axis + 0.001, Zcudf, 0.8, label = 'CuDF')
  
plt.xticks(X_axis, X)
plt.xlabel("Library")
plt.ylabel("Miliseconds(ms)")
plt.title("Measure performance based on CPU Time User")
plt.legend()
plt.show()
```

for the third component, which is using Isna Function. CuDF gains 4530ms, meanwhile Pandas gains 29.6 ms. This shows, CuDF has lesser the amount of time the processor spends in running this component.
